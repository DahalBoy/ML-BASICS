```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_student=pd.read_csv('Data/student_scores.csv')
display(data_student.head(6))
```

```python
X = data_student['Hours']
Y = data_student['Scores']
'''
you can also solve by numpy array
'''
#X=np.array(data_student['Hours'])
#Y=np.array(data_student['Scores'])
#display(X)
#print(X) --------shows all values in X
plt.scatter(X,Y)
plt.xlabel('Hours')
plt.ylabel('Scores')
```

```python
def compute_m_C(X,Y):
    m = 20
    c = 10
    a = 0.00001
    iteration = 100000
    for i in range(iteration):
        for j in range(len(X)):
            y_pred = m*X[j] + c            
            m = m - 2*a*(y_pred - Y[j])*X[j]
            c = c - 2*a*(y_pred - Y[j])
    print(m,c)
    arr = [m,c]
    return arr

arr = compute_m_C(X,Y)
            


```

```python
#Y = m*X +cH

Y_pred=[]
def Y_pred_val():
    for i in range(len(X)):
        value= arr[0]*X[i]+arr[1]
        #print(value)
        Y_pred.append(value)
    return Y_pred

#Easy method
'''Y_pred=[slope*i+intercept for i in X]'''
print(Y_pred)
Y_pred_val()
plt.plot(X,Y_pred, color='black')
plt.scatter(X,Y)
```

```python
X_input = float(input('Enter how many hour you read')) #X_inference
Predicted_Score = 0
def predict_score(X_input):
    Predicted_Score = arr[0]*X_input+arr[1]
    return Predicted_Score

Predicted_Score = predict_score(X_input)
if Predicted_Score > 100:
    Predicted_Score = 100
print(Predicted_Score)
```
